#Functions
-1. Difference Between a Function and a Method in Python:

A function in Python is a block of reusable code that performs a specific task. It is defined using the def keyword and can exist independently, outside of any class or object.

A method is a function that is associated with an object. It is called on an object and often acts on that objectâ€™s data. Methods are defined within a class and implicitly take the instance (self) as the first parameter.

In short: All methods are functions, but not all functions are methods. Methods are functions that belong to objects (especially classes).



Example:

```python
# Function
def greet(name):
    return "Hello " + name

print(greet("Taha"))

# Method
name = "Taha"
print(name.upper())  # upper() is a method of string object
```

-2. Function Arguments and Parameters in Python
Parameter: Variable listed in function definition.They are the names used in the function definition. They act as placeholders for the values that will be passed into the function.

Argument: are the actual values supplied to the function when it is called.

When a function is invoked, the arguments provided are matched with the parameters defined. Python supports different types of arguments: positional, keyword, default, and variable-length (*args, **kwargs).



Example:

```python
def add(x, y):  # x and y are parameters
    return x + y

result = add(5, 3)  # 5 and 3 are arguments
print(result)
```

-3. Ways to Define and Call a Function in Python
Regular function:

Python supports several ways to define and call functions:

Standard Function Definition: Using the def keyword.

Function with Default Arguments: Providing default values to parameters.

Keyword Arguments: Calling a function using parameter names.

Variable-length Arguments: Using *args for positional and **kwargs for keyword arguments.

Lambda Functions: Anonymous, short functions.

Recursive Functions: Functions that call themselves.

Nested Functions: Functions defined inside another function.

Calling a function can be done by its name followed by parentheses, optionally passing required arguments.





Example:

```python
def greet(name="Guest"):  # default argument
    print("Hello", name)

greet()  # calls with default
greet("Taha")  # calls with argument
```

- 4. Purpose of return Statement
The return statement is used to send a result back to the function caller.

It ends the execution of a function and optionally passes back a value.

A function without a return statement returns None by default.

It allows data to flow out of a function and be reused elsewhere in the program.

Example:

```python
def square(x):
    return x * x

print(square(4))  # returns 16
```

-5. Iterators vs Iterables:

An iterable is any Python object capable of returning its elements one at a time. Examples include lists, strings, tuples, and sets. Iterables implement the __iter__() method.

An iterator is an object that represents a stream of data; it returns one element at a time using the __next__() method. An iterator is created from an iterable using the iter() function.

Key difference: Iterables can be looped over, but only iterators maintain their current position during iteration.

Example:

```python
my_list = [1, 2, 3]
it = iter(my_list)  # create iterator

print(next(it))  # 1
print(next(it))  # 2
```

-6. Generators in Python
Generators are functions that yield items one at a time using the yield keyword.
Generators are a type of iterable like lists or tuples, but instead of holding all values in memory, they yield items one at a time using the yield keyword.

They are defined like regular functions but use yield instead of return.

When called, a generator function returns a generator object that can be iterated using next().

Each call to yield pauses the function, preserving its state for the next call.


Example:

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

for num in count_up_to(3):
    print(num)  # prints 1 2 3
```

-7. Advantages of Generators Over Regular Functions:

Memory Efficiency: Generators compute values lazily (one at a time), which is ideal for handling large datasets or infinite sequences.

Faster Performance: Especially when not all items are needed at once.

Simpler Code: Generator functions automatically manage the state, making them more concise than manually implemented iterator classes.

No Need for Entire Collection: Useful when you're processing data streams or large files.
Example:

```python
def gen():
    for i in range(1000000):
        yield i

g = gen()
print(next(g))  # doesn't load all values, just generates next one
```

-8. Lambda Function
A lambda function is a small, anonymous function defined using the lambda keyword.

It can take any number of arguments but only has one expression.

Lambda functions are often used when a short function is required temporarily for operations like sorting, mapping, or filtering.

They are not suited for complex operations and are typically used for short, inline functions

Example:

```python
square = lambda x: x * x
print(square(5))  # Output: 25
```

- 9. Purpose and Usage of map():

map() is a built-in function that applies a specified function to all items in an iterable (like a list or tuple).

It returns a map object, which is an iterator that can be converted into a list or other iterable.

map() is useful for transforming data without writing explicit loops.

It helps in applying a transformation function across a sequence of items concisely and efficiently.




Example:

```python
nums = [1, 2, 3]
squared = map(lambda x: x*x, nums)
print(list(squared))  # [1, 4, 9]
```

10. Difference Between map(), reduce(), and filter():

map() applies a function to every item in an iterable and returns a new iterable of the same length.

filter() applies a function that returns a Boolean (True/False) and only includes items for which the function returns True.

reduce() applies a rolling computation to sequential pairs of items in an iterable to reduce it to a single value. It is not built-in and must be imported from the functools module.

Key differences:

map() transforms elements,

filter() selects elements,

reduce() aggregates elements.



Example:

```python
from functools import reduce

nums = [1, 2, 3, 4]

# map
squared = list(map(lambda x: x**2, nums))

# filter
evens = list(filter(lambda x: x % 2 == 0, nums))

# reduce
product = reduce(lambda x, y: x * y, nums)

print(squared)  # [1, 4, 9, 16]
print(evens)    # [2, 4]
print(product)  # 24
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print("4. Prime numbers from 1 to 200:")
print([num for num in range(1, 201) if is_prime(num)])
```

```python
#5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms:
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

print("5. Fibonacci sequence:")
for num in FibonacciIterator(10):  # First 10 Fibonacci numbers
    print(num, end=' ')
print()
```

```python
#6Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exp):
    for exp in range(max_exp + 1):
        yield 2 ** exp

print("6. Powers of 2 up to exponent 5:")
for val in powers_of_two(5):
    print(val, end=' ')
print()

```

```python
#7 Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

# To use this function, create a test file first
with open("sample.txt", "w") as f:
    f.write("Line 1\nLine 2\nLine 3")

print("7. Reading lines from file:")
for line in read_file_lines("sample.txt"):
    print(line)
```

```python
#10Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(s):
    return ''.join(filter(lambda ch: ch.lower() not in 'aeiou', s))

print("10. String without vowels:", remove_vowels("Beautiful Day"))
# Output: "Btfl Dy"
```

```python
#12Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
#product of the price per item and the quantity. The product should be increased by 10,- â‚¬ if the value of the
#order is smaller than 100,00 â‚¬.
def calculate_order_totals(orders):
    result = []
    for order in orders:
        order_number, price_per_item, quantity = order
        total = price_per_item * quantity
        if total < 100:
            total += 10
        result.append((order_number, total))
    return result

# Sample data: (order_number, price_per_item, quantity)
orders = [
    (1, 10.0, 5),     # 50.0 â†’ +10 â†’ 60.0
    (2, 25.0, 4),     # 100.0 â†’ no change
    (3, 30.0, 2),     # 60.0 â†’ +10 â†’ 70.0
    (4, 50.0, 3),     # 150.0 â†’ no change
]

# Call the function and print the result
result = calculate_order_totals(orders)
print(result)
```

```python
#13Write a Python program using lambda and map.

# Sample list of orders: (order_number, price_per_item, quantity)
orders = [
    (1, 10.0, 5),    # 50 â†’ +10 â†’ 60
    (2, 25.0, 4),    # 100 â†’ no change
    (3, 30.0, 2),    # 60 â†’ +10 â†’ 70
    (4, 50.0, 3)     # 150 â†’ no change
]

# Use map and lambda to compute final values
result = list(map(
    lambda order: (
        order[0],
        order[1] * order[2] + (10 if order[1] * order[2] < 100 else 0)
    ),
    orders
))

# Print the result
print(result)
```
